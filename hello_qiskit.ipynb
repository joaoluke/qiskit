```python
import qiskit
```

```python
q_reg = qiskit.QuantumRegister(2)
c_reg = qiskit.ClassicalRegister(2)

circuit = qiskit.QuantumCircuit(q_reg, c_reg)
```

```python
circuit.draw()
```

```python
circuit.x(0)
circuit.draw()
```

```python
circuit.h(1)
circuit.cx(0, 1)

circuit.draw()
```

```python
circuit.measure(0,0)
circuit.measure(1,1)

circuit.draw('mpl')
```

```python
sim = qiskit.Aer.get_backend('qasm_simulator')
job = qiskit.execute(circuit, sim, shots=1024)
qiskit.visualization.plot_histogram(job.result().get_counts())
```

```python
c_reg = qiskit.ClassicalRegister(1)
q_reg = qiskit.QuantumRegister(1)

circuit = qiskit.QuantumCircuit(q_reg, c_reg)

circuit.x(0)
circuit.draw('mpl')
```

```python
qiskit.quantum_info.Statevector.from_instruction(circuit)
```
